```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report
```

```python
#Load dataset
X, y = load_iris(return_X_y=True)
```

```python
#Define feature description
X_desc = load_iris().feature_names
```

```python
#Define target description
y_desc = load_iris().target_names
```

```python
#Names and targets
y_desc, X_desc
```

```python
#Split the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=60)
X_train, y_train
```

```python
#Train the Naive Bayes Classifier
gnb = GaussianNB()
gnb.fit(X_train, y_train)
```

```python
#Use trained classifier gnb to make pred. on test data
y_pred = gnb.predict(X_test)
```

```python
#Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
#Display classification report
print(classification_report(y_test, y_pred, target_names=y_desc))
```

```python
print("Number of mislabeled points out of a total %d points: %d" % (X_test.shape[0], (y_test != y_pred).sum()))
```
